# 3) Write a program to simulate the digital-digital conversion techniques NRZ, NRZ-L, Manchester and Differential Manchester

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Unipolar NRZ

```python
def NRZ(data):

    x=np.arange(0, len(data), 0.01)
    y=np.empty(0)

    for bit in data:
        if bit=='0':
            y=np.concatenate((y, np.zeros(100)))
        elif bit=='1':
            y=np.concatenate((y, np.ones(100)))
            
    return x,y
```

```python
def plot_UNRZ(data):

    x,y=NRZ(data)
    plt.figure(figsize=(15, 4))
    plt.plot(x,y,linewidth=1)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Unipolar NRZ')
    plt.show()


data=input("Input Binary Data Stream : ")


plot_UNRZ(data)
```

# Polar NRZ-L

```python
def polar_NRZL(data):
    
    x=np.arange(0, len(data), 0.01)
    y=np.empty(0)

    for bit in data:
        if bit=="0":
                y=np.concatenate((y, np.ones(100)*1))
        elif bit=="1":
                y=np.concatenate((y, np.ones(100)*-1))
    
    return x,y
```

```python
def plot_PNRZ(data):
    x,y=polar_NRZL(data)
    plt.figure(figsize=(15, 4))
    plt.plot(x,y,linewidth=1)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('NRZ-L')
    plt.show()
    
data=input("Input Binary Data Stream : ")
plot_PNRZ(data)
```

# Manchester

```python
def Manchester(data: list):
    
    plot_initializer(len(data))
    plt.title("Manchester")
   
    for i in range(len(data)):
        if data[i] == 0:
            mid_x = round((i+(i+1))/2, 2)
            plt.plot([i,mid_x, mid_x, i+1], [0.75, 0.75, 0.25, 0.25], color="blue", linewidth=1)
        else:
            mid_x = round((i+(i+1))/2, 2)
            plt.plot([i,mid_x, mid_x, i+1], [0.25, 0.25,0.75, 0.75], color="blue", linewidth=1)
        if data[i] == data[i-1] and i>0:
                plt.plot([i,i], [0.75,0.25], color="blue", linewidth=1)
       
        plt.text(round((i+(i+1))/2, 2),0.15, s=str(data[i]), color="blue")
    
    plt.show()
    
# Initializing The Graph
def plot_initializer(xli: int):
    
   
    plt.figure(figsize=(15, 4))
    plt.xlim((0, xli))
    plt.ylim((0,1))
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.yticks([0.25, 0.50, 0.75],["-v","0", "+v"])
    plt.xticks([xli//2], [" "])  
    
if __name__ == '__main__':
    
    bit_data = [0, 1, 1, 0, 0, 1, 1]
    
Manchester(bit_data)    
```

# Differential Manchester

```python
def Diff_Manchester(data: list):

    plot_initializer(len(data))
    plt.title("Differential Manchester Encoding")
   
    stop = 0.75  #Setting 0.75 to start from positive side

    for i in range(len(data)):
        if data[i] == 0:
            if stop == 0.75:
                mid_x = round((i+(i+1))/2, 2)
                plt.plot([i,i,mid_x,mid_x, i+1], [stop, stop-0.50,stop-0.50,stop,stop], color="blue", linewidth=1)
            else:
                mid_x = round((i+(i+1))/2, 2)
                plt.plot([i,i,mid_x,mid_x, i+1], [stop, stop+0.50,stop+0.50,stop,stop], color="blue", linewidth=1)
        else:
            if stop == 0.75:
                mid_x = round((i+(i+1))/2, 2)
                plt.plot([i,mid_x,mid_x, i+1], [stop, stop, stop-0.50,stop-0.50], color="blue",linewidth=1)
                stop -= 0.50
            else:
                mid_x = round((i+(i+1))/2, 2)
                plt.plot([i,mid_x,mid_x, i+1], [stop,stop, stop+0.50,stop+0.50], color="blue", linewidth=1)
                stop += 0.50

        plt.text(round((i+(i+1))/2, 2),0.15, s=str(data[i]), color="blue")
    
    plt.show()
    
    
# Initializing The Graph
def plot_initializer(xli: int):
    
   
    plt.figure(figsize=(15, 4))
    plt.xlim((0, xli))
    plt.ylim((0,1))
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.yticks([0.25, 0.50, 0.75],["-v","0", "+v"])
    plt.xticks([xli//2], [" "]) 
    

if __name__ == '__main__':
    
    bit_data = [0, 1, 1, 0, 0, 1, 1]

    Diff_Manchester(bit_data)    
```
